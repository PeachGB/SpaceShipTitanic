```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split,cross_val_score

import matplotlib.pyplot as plt 

```

```python
df = pd.read_csv("/kaggle/input/spaceship-titanic/train.csv")
df_test = pd.read_csv('/kaggle/input/spaceship-titanic/test.csv')
```

```python
df.drop('Name',axis=1, inplace=True)
df_test.drop('Name',axis=1, inplace=True)
```

```python
numCols =[xs for xs in df if df[xs].dtype == "float64"]

for xs in numCols:
    mean = df[xs].mean()
    df[xs] = df[xs].fillna(mean)
    df_test[xs] = df_test[xs].fillna(mean)
```

```python
df.dropna(inplace = True)
labelCols= ['CryoSleep','VIP','Transported','Cabin']
encoder = LabelEncoder()
for i in labelCols:
    df[i] = encoder.fit_transform(df[i])
for col in [col for col in labelCols if col != 'Transported']:
    df_test[col] = encoder.fit_transform(df_test[col])
```

```python
cats = ['HomePlanet','Destination']

df = pd.get_dummies(df,columns = cats)
df_test = pd.get_dummies(df_test,columns = cats)

```

```python
X = df.drop('Transported',axis=1)
y = df['Transported']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33, random_state=42)
model = RandomForestClassifier(max_depth=8, random_state = 42)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
y_final = model.predict(df_test)
```

```python
y_id = df_test['PassengerId']
y_final = y_final.astype(bool)

final_df = pd.DataFrame({'PassengerId': y_id,'Transported':y_final})

final_df.to_csv('sample.csv',index=False)
```
